```python
# Visualizing the distribution of prices using a histogram
plt.figure(figsize=(12, 5))
sns.histplot(df['price'], color='b', kde=True)
plt.title('Distribution of Prices', fontsize=15)
plt.show()
```

```python
def detect_zscore_outliers(df, column, threshold=3):
    mean = df[column].mean()
    std_dev = df[column].std()
    z_scores = (df[column] - mean) / std_dev
    upper_bound = mean + threshold * std_dev
    outliers = df[z_scores.abs() > threshold]
    outlier_count = len(outliers)
    outlier_percentage = (outlier_count / len(df)) * 100
    
    print("Number of outliers detected using Z-scores method:", outlier_count)
    print("Percentage of outliers:", round(outlier_percentage, 2), "%")
    print("Upper bound for outliers:", round(upper_bound, 2))
    
    return detect_zscore_outliers

detect_zscore_outliers(df, 'price')
```

```python
# Handling outliers in 'price' and 'minimum_nights'
df = df[(df['price'] <= 1042.03)]
```

- Using the Z-score method, outliers were detected in the 'price' column.
- Listings with prices exceeding the upper bound (£1042.03) were removed.

```python
# Display shape of the dataset after removing outliers
print("Shape of the dataset after removing outliers:", df.shape)
```

```python
plt.figure(figsize=(12, 5))
sns.histplot(df['price'], color='b', kde=True)
plt.title('Distribution of Prices', fontsize=15)
plt.show()
```

```python
# Distribution of Listings Across Neighborhood Groups
plt.figure(figsize=(12, 8))
ax = sns.countplot(x='neighbourhood_group', data=df, palette='deep' )

# Adding annotations
total = len(df)
counts = df['neighbourhood_group'].value_counts()
ratios = counts / total * 100
for i, (count, ratio) in enumerate(zip(counts, ratios)):
    plt.text(0.95, 0.95 - i*0.05, f'{counts.index[i]}: {count} ({ratio:.2f}%)',
             ha='right', va='top', transform=ax.transAxes)
    
plt.title('Distribution of Listings Across Neighborhood Groups')
plt.show()
```

The data highlights significant discrepancies in Airbnb listing distribution across Greater Manchester's neighborhood groups. Manchester and Salford host the highest number of listings, likely due to their central locations and cultural appeal. Conversely, Rochdale exhibits the lowest number of listings, potentially influenced by lower demand or less developed tourism infrastructure. This distribution reflects varying levels of area attractiveness and demand within Greater Manchester.

```python
# Density and distribution of room_type
plt.figure(figsize=(12, 8))
ax = sns.countplot(x='room_type', data=df, palette="deep")

# Adding annotations
total = len(df)
counts = df['room_type'].value_counts()
ratios = counts / total * 100
for i, (count, ratio) in enumerate(zip(counts, ratios)):
    plt.text(0.95, 0.95 - i*0.05, f'{counts.index[i]}: {count} ({ratio:.2f}%)',
             ha='right', va='top', transform=ax.transAxes)
    
plt.title('Density and distribution of room type')
plt.show()
```

The analysis reveals a predominant preference for entire home/apartment listings, comprising 62.62% of the total. Private rooms follow with 36.15%, while shared rooms and hotel rooms represent smaller fractions at 1.01% and 0.21%, respectively. This distribution underscores the popularity of private accommodations among Airbnb users in Greater Manchester.

```python
# Mean Price of Listings for Each Neighbourhood Group
mean_prices = df.groupby('neighbourhood_group')['price'].mean().reset_index()
plt.figure(figsize=(12, 8))
sns.barplot(x="neighbourhood_group", y="price", data=mean_prices, palette="deep")

for index, row in mean_prices.iterrows():
    plt.text(index, row['price'] + 1, f"${row['price']:.2f}", ha='center')
plt.title('Mean Price of Listings for Each Neighbourhood Group')
plt.show()
```

The mean price of listings varies across different neighborhoods in Greater Manchester. Oldham has the highest average price at 141.00, followed by Manchester at 126.01, and Salford at 120.55. Conversely, Tameside has the lowest average price at 97.44, with Rochdale and Bolton also offering relatively lower average prices at 106.56 and 105.76, respectively. These price differences may reflect varying demand, amenities, and neighborhood characteristics.

```python
# Room Type of Listings for Each Neighbourhood Group
mean_prices = df.groupby('room_type')['price'].mean().reset_index()
plt.figure(figsize=(12, 8))
sns.barplot(x="room_type", y="price", data=mean_prices, palette="deep")

for index, row in mean_prices.iterrows():
    plt.text(index, row['price'] + 2, f"${row['price']:.2f}", ha='center')
    
plt.title('Average prices according to room type')
plt.show()
```

As we can see here, the hotel rooms are the highest in price despite their small number, and the shared room is one of the lowest in price and also one of the least common. The private room is slightly higher than the shared room, and the full house is higher than both.
